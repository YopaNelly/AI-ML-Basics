```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Boxplot.csv')
```

```python
# draw boxplot for each teaching method in different frames horizontally
teaching_methods = df["Teaching Method"].unique()
num_methods = len(teaching_methods)

fig, axes = plt.subplots(1, num_methods, figsize=(15, 5))

for i, method in enumerate(teaching_methods):
    sns.boxplot(y="Score", data=df[df["Teaching Method"] == method], ax=axes[i])
    axes[i].set_title(method)

plt.tight_layout()
plt.show()
```

```python
# draw boxplot foe each teaching method in the same frame

plt.figure(figsize=(10, 5))
sns.boxplot(x="Teaching Method", y="Score", data=df)
plt.show()
```

```python
# Draw boxplot for each teaching method in the same frame with individual data points (swarm plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x="Teaching Method", y="Score", data=df)
sns.swarmplot(x="Teaching Method", y="Score", data=df, color=".25")  # Add swarm plot
plt.title("Scores by Teaching Method with Individual Data Points")
plt.show()
```

```python
# Plots horizontal boxplot
sns.boxplot(data=df, orient='h', palette='Set2')
plt.title('Colored Boxplot')
plt.show()
```
