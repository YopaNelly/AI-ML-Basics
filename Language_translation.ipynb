```python
import torch
from transformers import MarianMTModel, MarianTokenizer, Trainer, TrainingArguments
```

```python
# Step 1: Load the Pretrained Model and Tokenizer
model_name = 'Helsinki-NLP/opus-mt-fr-en'  # Pretrained model for French to English
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
# Step 2: Prepare Dataset
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your French-Medumba dataset
data = pd.read_csv('translations.csv')
train_data, val_data = train_test_split(data, test_size=0.1)
```

```python
# Step 3: Tokenize Dataset
# Helper function to tokenize data
def tokenize_function(examples):
    # Tokenize inputs (French) and targets (Medumba)
    inputs = tokenizer(examples['French Term'], truncation=True, padding=True, max_length=128)
    targets = tokenizer(examples['Translation'], truncation=True, padding=True, max_length=128)

    # Define input and output labels
    inputs['labels'] = targets['input_ids']
    return inputs
```

```python
# Tokenize training and validation data
train_encodings = train_data.apply(tokenize_function, axis=1)
val_encodings = val_data.apply(tokenize_function, axis=1)
```

```python
# Step 4: Create PyTorch Dataset Class
from torch.utils.data import Dataset

class TranslationDataset(Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}

    def __len__(self):
        return len(self.encodings['input_ids'])
```

```python
# Create PyTorch datasets
train_dataset = TranslationDataset(train_encodings)
val_dataset = TranslationDataset(val_encodings)
```

```python

# Helper function to tokenize data
def tokenize_function(examples):
    # Tokenize French inputs and Medumba targets
    inputs = tokenizer(examples['French Term'], truncation=True, padding="max_length", max_length=128)
    targets = tokenizer(examples['Translation'], truncation=True, padding="max_length", max_length=128)

    # Assign 'input_ids' and 'attention_mask' for inputs and 'labels' for targets
    inputs['labels'] = targets['input_ids']
    return inputs

# Create a PyTorch Dataset class
class TranslationDataset(Dataset):
    def __init__(self, df):
        self.encodings = df.apply(tokenize_function, axis=1).tolist()

    def __getitem__(self, idx):
        # Extract the encoding and convert values to torch tensors
        encoding = self.encodings[idx]
        return {key: torch.tensor(val) for key, val in encoding.items()}

    def __len__(self):
        return len(self.encodings)

# Convert the training and validation data to the TranslationDataset format
train_dataset = TranslationDataset(train_data)
val_dataset = TranslationDataset(val_data)

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",        # Keep this as is for now
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=1,
    weight_decay=0.01,
)

# Create Trainer instance
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Train the model
trainer.train()

# Save the fine-tuned model
model.save_pretrained("./fine_tuned_french_medumba")
tokenizer.save_pretrained("./fine_tuned_french_medumba")

# Test inference with the fine-tuned model
fine_tuned_model = MarianMTModel.from_pretrained("./fine_tuned_french_medumba")
fine_tuned_tokenizer = MarianTokenizer.from_pretrained("./fine_tuned_french_medumba")

french_sentence = "Bonjour, comment ça va?"
inputs = fine_tuned_tokenizer(french_sentence, return_tensors="pt")
outputs = fine_tuned_model.generate(**inputs)

# Decode the translation
translation = fine_tuned_tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Translated to Medumba:", translation)
```
