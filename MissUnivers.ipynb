```python
# prompt: convert missUniWor.xlsx  to dataframe

import pandas as pd

try:
  df = pd.read_excel('missUniWor.xlsx')
  print(df.head()) # Print the first few rows to verify
except FileNotFoundError:
  print("Error: 'missUniWor.xlsx' not found. Check the file name and path.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
# drop null rows

df.dropna(inplace=True)
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(x='Country', data=df, order = df['Country'].value_counts().index)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Frequency of Countries')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.scatterplot(x='Year', y='Age', data=df) # Use seaborn for the plot

# Customize the plot
plt.title('Scatter Plot of Age vs. Year')
plt.xlabel('Year')
plt.ylabel('Age')
plt.grid(True)  # Add a grid for better readability

# Show the plot
plt.show()
```
