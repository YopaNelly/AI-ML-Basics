```python
import networkx as nx
import random
from networkx.algorithms.community.quality import modularity

def lpam(G):
    # Step 1: Initialize each node with its own label
    labels = {node: node for node in G.nodes()}
    m = G.number_of_edges()

    def get_modularity(G, labels):
        # Convert labels to communities for modularity calculation
        communities = {}
        for node, label in labels.items():
            communities.setdefault(label, set()).add(node)
        return modularity(G, communities.values())

    changed = True
    while changed:
        changed = False
        nodes = list(G.nodes())
        random.shuffle(nodes)

        for node in nodes:
            best_label = labels[node]
            max_modularity = get_modularity(G, labels)

            # Try all neighbor labels
            neighbor_labels = set(labels[neighbor] for neighbor in G.neighbors(node))
            for label in neighbor_labels:
                old_label = labels[node]
                labels[node] = label
                new_modularity = get_modularity(G, labels)
                if new_modularity > max_modularity:
                    best_label = label
                    max_modularity = new_modularity
                labels[node] = old_label  # revert temporarily

            if labels[node] != best_label:
                labels[node] = best_label
                changed = True

    # Collect communities
    communities = {}
    for node, label in labels.items():
        communities.setdefault(label, set()).add(node)

    return list(communities.values())
```

```python
import matplotlib.pyplot as plt

def draw_communities(G, communities):
    pos = nx.spring_layout(G)
    colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']
    for i, comm in enumerate(communities):
        nx.draw_networkx_nodes(G, pos, nodelist=comm, node_color=colors[i % len(colors)], label=f"Community {i+1}")
    nx.draw_networkx_edges(G, pos)
    nx.draw_networkx_labels(G, pos)
    plt.legend()
    plt.show()

draw_communities(G, communities)
```

```python
import community
partition = community.best_partition(G)

# Group nodes by community
from collections import defaultdict
comms = defaultdict(list)
for node, comm_id in partition.items():
    comms[comm_id].append(node)

for i, members in comms.items():
    print(f"Community {i+1}: {sorted(members)}")
```
