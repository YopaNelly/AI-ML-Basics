```python
import os
import pandas as pd
import numpy as np

# Directory containing the extracted dataset
root_dir = "data10"  # change to your root folder path
output_dir = "converted_csv"
os.makedirs(output_dir, exist_ok=True)

for dirpath, _, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith("conn.log.labeled"):
            file_path = os.path.join(dirpath, filename)
            print(f"📄 Processing: {file_path}")

            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    lines = f.readlines()

                # Extract the header line that starts with '#fields'
                field_line = next(line for line in lines if line.startswith('#fields'))
                fields = field_line.strip().split('\t')[1:]  # remove the '#fields' tag

                # Extract data lines (ignore comment lines starting with '#')
                data_lines = [line.strip() for line in lines if not line.startswith('#') and line.strip()]

                # Split each data line by tabs
                data = [line.split('\t') for line in data_lines]

                # Filter out malformed rows that do not match the header length
                data = [row for row in data if len(row) == len(fields)]

                # Create DataFrame
                df = pd.DataFrame(data, columns=fields)

                # Replace '-' with NaN
                df.replace('-', np.nan, inplace=True)

                # Use the name of the parent folder as CSV filename (replace slashes if any)
                parent_folder = os.path.basename(dirpath)
                output_file = f"{parent_folder}.csv"
                output_path = os.path.join(output_dir, output_file)

                # Save DataFrame to CSV
                df.to_csv(output_path, index=False)
                print(f"✅ Saved CSV: {output_path}")

            except Exception as e:
                print(f"❌ Failed to process {file_path}: {e}")
```

```python
import os
import pandas as pd
import numpy as np

root_dir = "data10"  # Your dataset root folder
output_dir = "converted_csv"
os.makedirs(output_dir, exist_ok=True)

for dirpath, _, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith("conn.log.labeled"):
            file_path = os.path.join(dirpath, filename)
            print(f"📄 Processing: {file_path}")

            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    lines = f.readlines()

                field_line = next(line for line in lines if line.startswith('#fields'))
                fields = field_line.strip().split('\t')[1:]

                data_lines = [line.strip() for line in lines if not line.startswith('#') and line.strip()]
                data = [line.split('\t') for line in data_lines]
                data = [row for row in data if len(row) == len(fields)]

                df = pd.DataFrame(data, columns=fields)
                df.replace('-', np.nan, inplace=True)

                # Create a unique filename based on relative path
                rel_path = os.path.relpath(file_path, root_dir)
                csv_name = rel_path.replace('/', '_').replace('\\', '_') + '.csv'

                output_path = os.path.join(output_dir, csv_name)
                df.to_csv(output_path, index=False)

                print(f"✅ Saved CSV: {output_path}")

            except Exception as e:
                print(f"❌ Failed to process {file_path}: {e}")
```

```python
# Function to find highly correlated independent columns
def find_highly_correlated_independent_cols(df, correlation_threshold=0.9):
    """
    Finds pairs of highly correlated numerical columns.

    Args:
        df (pd.DataFrame): The input DataFrame.
        correlation_threshold (float): The absolute correlation threshold
                                       above which columns are considered highly correlated.

    Returns:
        list: A list of tuples, where each tuple contains a pair of column names
              that are highly correlated.
    """
    # Calculate the correlation matrix for numerical columns
    numerical_df = df.select_dtypes(include=np.number)
    correlation_matrix = numerical_df.corr().abs()

    # Select upper triangle of correlation matrix
    upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

    # Find features with correlation greater than the threshold
    highly_correlated_pairs = [(column, row) for column in upper.columns for row in upper.index if upper.loc[row, column] >= correlation_threshold]

    return highly_correlated_pairs

# Find highly correlated independent columns
highly_correlated_cols = find_highly_correlated_independent_cols(combined_df_cleaned, correlation_threshold=0.9)

print("\nHighly Correlated Independent Columns (Absolute Correlation >= 0.9):")
if highly_correlated_cols:
    for col1, col2 in highly_correlated_cols:
        print(f"  - {col1} and {col2}")
else:
    print("No highly correlated independent columns found above the threshold.")
```
